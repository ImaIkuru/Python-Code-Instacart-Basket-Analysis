## 1 Importing libraries

## 2 Visualizations

### 01 orders_day_of_week bar chart

```python
#bar chart for days of the week, plotting frequency
bar_dow= df['orders_day_of_week'].value_counts().plot.bar(color='#54a868',xlabel='days of the week',ylabel='orders') 
```

### 02 prices Histogram and scatterplot

```python
#creating histogram of the prices column and bin size of 25
hist_prices=df['prices'].plot.hist(bins = 70, color='#54a868')
```

```python
#creating scatterplot for the prices column, x-axis and y-axis are price
scatter_prices=sns.scatterplot(x = 'prices', y = 'prices',data= df,color='#54a868')
```

### 03 orders_day_of_week and prices  line chart with sampling

### Creating random samples

```python
#creating random samples with 70/30 split
#the seed represents a random but reproducable pattern
#dev stores boolean values for if value in array is less than or greater than 0.7
#random created random numbers and rand ensures uniform distribution is followed
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
#using list to split into two
#storing 70%
big = df[dev]
```

```python
#storing 30%
small = df[~dev]
```

```python
#accuracy check
len(df)
```

```python
len(big)+len(small)
```

```python
#reducing dataframe into only necessary columns for faster processing
df_2 = small[['orders_day_of_week','prices']]
```

### Creating line chart

```python
line_dow = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices',color='#54a868')
```

### 04  Order_hour_of_day histogram

```python
#creating histogram of the order_hour_of_day column and bin size of 10
hist_hod=df['order_hour_of_day'].plot.hist(bins = 24,color='#54a868')
```

#### The histogram shows how many orders are made each hour of the day, the fewest orders are made in the early morning and late nights while the most orders are made in the daytime

### 05 Loyalty flag bar chart

```python
#bar chart for loyalty flag, plotting frequency
bar_loyalty=df['loyalty_flag'].value_counts().plot.bar(color='#54a868')
```

#### The bar chart shows how many orders were made by customers in each of the loyalty groups, the regular customers have made the most purchases

### 06 Line chart for hourly prices

```python
#using small random sample to create version of dataframe with only relevant columns
df_3 = small[['order_hour_of_day','prices']]
```

```python
#creating line chart from dataframe
#setting chart size
sns.set(rc = {'figure.figsize':(20,8)})
line_hourly_prices= sns.lineplot(data = df_3, x = 'order_hour_of_day',y = 'prices',color='#54a868')
```

#### The line chart shows the prices of orders made for each hour of the day, the most expensive orders are made earlier in the day

### 07 Line chart for numher of dependents vs age

```python
#creating scatterplot for the age and income columns
scatter_age_income=sns.scatterplot(x = 'age', y = 'income',data = small,color='#54a868')
```
